```python
import numpy as np

checkerboard = np.zeros((8, 8))

for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            checkerboard[i, j] = 1
        else:
            checkerboard[i, j] = 0

print(checkerboard)
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
inverse = np.linalg.inv(A)
mul = np.dot(A ,inverse)
print(mul)
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
product1 = np.dot(A,B)
product2 = np.dot(B,A)
comparison = product1 == product2
print(comparison)
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
product = np.dot(A,B)
transpose = np.transpose(product)
transpose1 = np.transpose(A)
transpose2 = np.transpose(B)
product1 = np.dot(transpose2,transpose1)
comparison = transpose == product1
print(comparison)
```
